```python
from scipy.stats import chi2_contingency
```

```python
Ratiodata=([50,142,131,70],[435,1523,1356,750])
Ratiodata
```

```python
stat, p, dof, expected = chi2_contingency(Ratiodata)
stat, p, dof, expected
```

```python
alpha = 0.05
print("p value is " + str(p))
if p <= alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (H0 holds true)')
```

```python
# Hence Thers is the samples are independent of each other .
```

```python
from scipy.stats import chi2_contingency
```

```python
Customerdata=([271,267,269,280],[29,33,31,20])
Customerdata
```

```python
stat, p, dof, expected = chi2_contingency(Customerdata)
stat, p, dof, expected
```

```python
alpha = 0.05
print("p value is " + str(p))
if p <= alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (H0 holds true)')
```

```python
# Hence as the p value is less than the alpha hence H0 is true ,hence the variables are independent to each other .
```

Fitting a Linear Regression Model


```python
import statsmodels.formula.api as smf
model = smf.ols("Salary~YearsExperience",data = data).fit()
model.summary()
```

```python
sns.regplot(x="YearsExperience", y="Salary", data=data);
```

```python
#Coefficients
model.params
```

```python
#t and p-Values
print(model.tvalues, '\n', model.pvalues)  
```

```python
#R squared values
(model.rsquared,model.rsquared_adj)
```

```python
#Predict for 4 & 3 years the salary hike
newdata=pd.Series([4,3])
newdata
```

```python
data_pred=pd.DataFrame(newdata,columns=['YearsExperience'])
model.predict(data_pred)
```

```python
predict = pd.DataFrame()
predict['YearsExperience'] = data.YearsExperience
predict['Salary'] = data.Salary
predict['Predict_Salary_Hike'] = pd.DataFrame(model.predict(predict.YearsExperience))
predict
```

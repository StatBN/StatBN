```python
user_book=book.pivot_table(index='User.ID',columns='Book.Title', values='Book.Rating').reset_index(drop=True)
user_book
```

```python
user_book.fillna(0,inplace=True)
user_book
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_book.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_booksim = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_booksim.index = book['User.ID'].unique()
user_booksim.columns = book['User.ID'].unique()
```

```python
user_booksim.iloc[0:5][0:10]
```

```python
np.fill_diagonal(user_sim,0)
user_booksim.iloc[0:5,0:5]
```

```python
#Most Similar Users
user_booksim.idxmax(axis=1)
```
